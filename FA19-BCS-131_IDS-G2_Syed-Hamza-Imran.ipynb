```python
df = pd.read_csv("Updated Dataset.csv")
```

```python
#  Plot a pie chart to show the ratio of male and female students.
male = df['Gender'].str.startswith('M').sum()
female = df['Gender'].str.startswith('F').sum()
gender = ["Male", "Female"]
data = [male, female]
plt.pie(data, labels=gender, autopct='%1.1f%%')
plt.title("Ratio of Male and Female Students")
plt.show()
```

```python
list = df[df['Gender'].str.startswith('M')]
maleList = list['CGPA'].tolist()
intervals = [2.0, 2.5, 3.0, 3.5, 4.0]
plt.hist(maleList, bins=intervals, ec='black')
plt.title("CGPA of all Male Students")
plt.xlabel("CGPA")
plt.ylabel("Number of Students")
plt.xticks(intervals)
plt.show()
```

```python
#  Plot the HSSC-1 marks of all male vs female students on a scatter plot.
randomlist = random.sample(range(0, 500), 21)
maleGrades = df[df['Gender'].str.startswith(
    'M')]["HSSC-1"].tolist()
femaleGrades = df[df['Gender'].str.startswith(
    'F')]["HSSC-1"].tolist()
maleGrades = maleGrades[:len(femaleGrades)]
plt.scatter(maleGrades, randomlist, marker='o', label="Male")
plt.scatter(femaleGrades, randomlist, marker='x', label="Female")
plt.xlabel('Marks Scored', fontsize=12)
plt.ylabel('Total Marks', fontsize=12)
plt.title("Marks of all Male vs Female Students")
plt.legend()
plt.show()
```

```python
# Plot the favorite colors of male vs female students on a bar chart.
maleFavColor = df[df['Gender'].str.startswith(
    'M')]["Favorite color"].count()
femaleFavColor = df[df['Gender'].str.startswith(
    'F')]["Favorite color"].count()
favColor = df.groupby(["Favorite color", "Gender"]
                           ).size().unstack(level=1).plot(kind='bar')
plt.xlabel("Male")
plt.ylabel("Female")
plt.show()
```

```python
#  Plot OF  line chart of students and their birth months.
birthMonth = df["Birthmonth"].tolist()
plt.plot(birthMonth, '-', label='birthmonth')
plt.xlabel("Students")
plt.ylabel("Birth Month")
plt.show()
```

```python
# 9. Create a correlation matrix between HSSC-1 and HSSC-2 marks and then plot on a heatmap.
marks = df.heatmap(df[["HSSC-1", "HSSC-2"]].corr(),
                   cmap='YlGnBu', annot=True)
plt.show()
```
